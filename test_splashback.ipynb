```python
import colossus.cosmology.cosmology as cosmo
import numpy as np
from colossus.lss import peaks
import matplotlib.pylab as plt
from colossus.halo import concentration, mass_so
from colossus.halo import splashback as sp
import utils

cosmol=cosmo.setCosmology('planck15')    #cosmology for colossus. By default hmf uses planck15
cosmo.setCurrent(cosmol)
```

```python
for M in [13,13.5]:
    z = np.arange(0,3,0.1)
    nu = peaks.peakHeight(10**M, z)
    RspR200m, mask = splashback.splashbackModel('RspR200m', nu200m = nu, z = z)
    plt.plot(z,RspR200m/RspR200m[0], label='Mh='+str(M))
plt.legend(frameon=False)
plt.ylabel('$R_{SP}/R_{SP}(z=0)$')
plt.xlabel('z')
```
